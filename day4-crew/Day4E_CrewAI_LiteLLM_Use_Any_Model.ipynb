# 1. CrewAI is Not the Model

This is a very important point for faculty.

CrewAI is not GPT.  
CrewAI is not Claude.  
CrewAI is not Gemini.  
CrewAI is not Ollama.

CrewAI is a framework that organizes Agents, Tasks, Tools, and Crews.

The actual intelligence comes from the LLM behind the Agent.

---

## Simple Mental Model

```text
LLM = Brain
CrewAI = Team coordination framework
Tools = External capabilities
Tasks = Work assignments
Crew = Group of Agents working together
```


# 2. Why Model Choice Matters

Different models have different strengths.

| Need | Possible Model Choice |
|---|---|
| Best quality reasoning | GPT-4 class / Claude class models |
| Lower cost | Smaller models |
| Faster response | Lightweight models |
| Local/private demo | Ollama models |
| Long documents | Models with larger context window |
| Classroom practice | Cost-effective models |

---

## Teaching Point

The framework may remain the same.

Only the model changes.


# 3. LiteLLM-Style Model Naming

CrewAI supports model names through LiteLLM-style naming.

Examples:

| Provider | Example Model Name |
|---|---|
| OpenAI | `gpt-4o-mini` |
| Anthropic | `claude-3-5-haiku` |
| Google | `gemini/gemini-2.0-flash` |
| Ollama | `ollama/llama3.2` |
| DeepSeek | `deepseek/deepseek-chat` |

The exact model names and API keys depend on your installed versions and provider setup.

For this FDP, we will keep the demo simple and use OpenAI by default.

````
from pathlib import Path
from dotenv import load_dotenv
import os

from crewai import Agent, Task, Crew, Process, LLM
````

# 5. Load Environment Variables

We will use the standard FDP root `.env` loading pattern.


````
current = Path.cwd().resolve()

for folder in [current] + list(current.parents):
    env = folder / ".env"
    if env.exists():
        load_dotenv(env)
        break

print("OpenAI API Key Loaded:", os.getenv("OPENAI_API_KEY") is not None)
````

# 6. Create a Default LLM

We will first create a simple OpenAI model configuration.

For classroom demos, a smaller fast model is often enough.

Example:

```python
LLM(model="gpt-4o-mini")
```

This keeps demos faster and cost-effective.


````
default_llm = LLM(
    model="gpt-4o-mini"
)

print("Default LLM configured.")
````

# 7. Build a Simple Agent Using This LLM

We will create a simple faculty assistant Agent and explicitly attach the LLM.


````
faculty_agent = Agent(
    role="Faculty AI Assistant",
    goal="Explain AI concepts in simple language for engineering faculty",
    backstory="""
    You are a helpful AI faculty assistant. You explain technical topics clearly,
    avoid jargon, and use classroom-friendly examples.
    """,
    llm=default_llm,
    verbose=True,
    allow_delegation=False,
)
````

# 8. Create a Simple Task

This task is intentionally small.

The purpose is not to build a large Crew.

The purpose is to show that the Agent is using the configured model.


````
explain_task = Task(
    description="""
    Explain why model selection matters in Agentic AI.

    Use:
    - one simple analogy
    - one classroom example
    - one enterprise example
    """,
    expected_output="A simple explanation suitable for engineering faculty.",
    agent=faculty_agent,
)
````

# 9. Create and Run the Crew

This is a single-Agent Crew.


````
model_demo_crew = Crew(
    agents=[faculty_agent],
    tasks=[explain_task],
    process=Process.sequential,
    verbose=True,
)

result = model_demo_crew.kickoff()

print(result)
````

# 10. What Did We Just Learn?

We explicitly configured the LLM used by the Agent.

```text
CrewAI Agent
   ↓
LLM Configuration
   ↓
Model Provider
   ↓
Response
```

The Agent structure stays the same.

Only the model can change.


# 11. Changing the Model

To switch models, we usually change the model name.

Example:

```python
LLM(model="gpt-4o-mini")
LLM(model="gpt-4o")
LLM(model="gemini/gemini-2.0-flash")
LLM(model="ollama/llama3.2")
```

Important:

Different providers require different API keys and local setup.

For example, Ollama requires the model to be running locally.

# 12. Classroom-Friendly Decision Guide

| Situation | Recommended Model Type |
|---|---|
| First classroom demo | Fast, low-cost model |
| Complex reasoning | Strong reasoning model |
| Long document analysis | Large context model |
| Privacy-sensitive demo | Local model through Ollama |
| Student practice | Low-cost model |
| Production enterprise workflow | Carefully benchmarked model |

---

## Important Teaching Point

Do not choose a model only because it is popular.

Choose based on:

- quality
- cost
- latency
- context size
- privacy
- integration support


# 13. Cost and Performance Discussion

For an FDP, cost and reliability matter.

A very powerful model may give better answers but cost more.

A smaller model may be cheaper and faster but may need better prompts.

---

## Simple Rule

Use the smallest model that gives acceptable quality for the task.

This is a very practical enterprise principle.


# 14. Enterprise Example

A company may use different models for different tasks.

```text
Customer FAQ Agent → Small fast model
Legal Review Agent → Strong reasoning model
Document Summary Agent → Large context model
Internal Prototype → Local Ollama model
Production Assistant → Approved enterprise model
```

This is called model selection strategy.


# 15. Academic Example

A college may use different models for different teaching tasks.

```text
Quiz generation → Small fast model
Research proposal review → Stronger model
Textbook summarization → Long-context model
Offline lab demo → Local model
```

Faculty should understand that one model is not always best for every job.


# 16. Mini Experiment

Change the model name in this cell if your environment supports it.

Then run the same task and compare:

- answer quality
- speed
- level of detail
- teaching usefulness


````
# Classroom experiment cell

experiment_llm = LLM(
    model="gpt-4o-mini"
)

experiment_agent = Agent(
    role="AI Concept Explainer",
    goal="Explain technical concepts in simple language",
    backstory="You are a faculty-friendly explainer for engineering education.",
    llm=experiment_llm,
    verbose=True,
    allow_delegation=False,
)

experiment_task = Task(
    description="""
    Explain CrewAI in 5 simple bullet points for engineering faculty.
    """,
    expected_output="Five simple bullet points.",
    agent=experiment_agent,
)

experiment_crew = Crew(
    agents=[experiment_agent],
    tasks=[experiment_task],
    process=Process.sequential,
    verbose=True,
)

experiment_result = experiment_crew.kickoff()

print(experiment_result)
````

# 19. Common Mistakes

Avoid these mistakes:

1. Assuming CrewAI itself is the model.
2. Using the strongest model for every task.
3. Ignoring cost during demos.
4. Ignoring latency during classroom sessions.
5. Forgetting provider-specific API keys.
6. Trying local models without setting up Ollama.
7. Not testing model output quality before the workshop.

---

## Teaching Tip

For live FDP demos, prioritize reliability.

It is better to use one stable model confidently than switch models and face setup issues during class.
