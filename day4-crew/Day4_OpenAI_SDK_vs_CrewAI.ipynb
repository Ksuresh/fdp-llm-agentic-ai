# 2. One Problem, Two Approaches

Imagine DSU wants an AI Teaching Assistant.

The assistant should:

- Create lesson plans
- Create quizzes
- Suggest lab exercises
- Recommend references
- Prepare classroom case studies

There are two ways to build this.


# 3. Approach 1: OpenAI Agents SDK

With the OpenAI Agents SDK, we write the orchestration logic ourselves.

```text
Faculty Request
        │
        ▼
Teaching Orchestrator
        │
 ┌──────┼────────────┐
 │      │            │
 ▼      ▼            ▼
Lesson Quiz       Demo
Agent  Agent      Agent
        │
        ▼
Combined Teaching Pack
```

The developer controls:

- Which Agent runs first
- Which Agent runs next
- Whether Agents run sequentially or in parallel
- How outputs are combined
- What happens if an Agent fails
- What quality checks are applied

This gives maximum control.


## Pause & Reflect

With OpenAI SDK:

- You control every Agent.
- You control routing.
- You control workflow.
- You control orchestration.
- You control error handling.

This is powerful, but it also means you write more orchestration code.


# 4. Approach 2: CrewAI

With CrewAI, we describe the team.

We define:

- Agents
- Roles
- Goals
- Tasks
- Expected outputs

CrewAI manages much of the collaboration.

```text
Faculty Request
        │
        ▼
Crew
        │
 ┌──────┼────────────┐
 │      │            │
 ▼      ▼            ▼
Lesson Quiz       Demo
Agent  Agent      Agent
        │
        ▼
Teaching Pack
```

The focus shifts from writing orchestration code to describing collaborative work.


# 5. Simple Analogy

## OpenAI SDK

OpenAI SDK is like building your own house.

You decide:

- room layout
- wiring
- plumbing
- materials
- architecture

It gives maximum flexibility and control.

---

## CrewAI

CrewAI is like using a modular building system.

You decide:

- purpose
- rooms
- roles
- final output

The framework helps arrange collaboration faster.

---

## Simple Summary

OpenAI SDK gives more control.

CrewAI gives faster multi-agent collaboration.


# 6. Side-by-Side Comparison

| Feature | OpenAI Agents SDK | CrewAI |
|---|---|---|
| Primary focus | Building Agents | Coordinating Agent teams |
| Coding required | Medium to High | Low to Medium |
| Workflow control | Manual | Built-in patterns |
| Multi-Agent support | Yes | Strong |
| Agent roles | Possible | Central idea |
| Task assignment | Manual | Natural and explicit |
| Orchestration | You write it | Framework helps |
| Rapid prototyping | Good | Excellent |
| Production flexibility | Excellent | Good to Excellent |
| Best for | Custom applications | Collaborative workflows |
| Beginner friendliness | Moderate | Easier |


# 7. When Should We Use OpenAI SDK?

Use OpenAI SDK when you need:

- Fine-grained control
- Production-grade APIs
- Custom routing
- Custom guardrails
- Custom tool execution
- Deep integration with enterprise systems
- Precise error handling
- Complex application logic

---

## Example

A banking customer support system where every decision must be logged, validated, and integrated with internal systems.

```text
Customer Request
   ↓
Authentication
   ↓
Policy Check
   ↓
Tool Call
   ↓
Audit Log
   ↓
Response
```

Here, OpenAI SDK is a strong choice because we may want complete control.


# 8. When Should We Use CrewAI?

Use CrewAI when you need:

- Quick prototypes
- Research demos
- Teaching examples
- Hackathons
- Collaborative AI workflows
- Role-based Agent teams
- Faster experimentation
- Less orchestration code

---

## Example

An AI Faculty Assistant that prepares:

- lesson plan
- quiz
- assignment
- reference list
- case study

This naturally feels like a team of specialists, which CrewAI handles well.


# 9. DSU Teaching Assistant Example

## OpenAI SDK Version

```text
Faculty Request
   ↓
Python Orchestrator
   ├── Lesson Agent
   ├── Quiz Agent
   ├── Demo Agent
   └── Reference Agent
   ↓
Combined Output
```

You explicitly write the orchestration.

---

## CrewAI Version

```text
Faculty Request
   ↓
Crew
   ├── Lesson Planner
   ├── Quiz Creator
   ├── Lab Instructor
   └── Reference Expert
   ↓
Teaching Pack
```

You define the team and tasks.

CrewAI coordinates the flow.


# 10. Important Clarification

CrewAI is **not another LLM**.

CrewAI can use LLMs such as:

- GPT
- Claude
- Gemini
- Ollama
- Other supported models

CrewAI is an orchestration framework.

It manages:

- Agents
- Roles
- Tasks
- Collaboration
- Process flow

The intelligence still comes from the underlying LLM.


# 11. Where Does LangGraph Fit?

You may also hear about LangGraph.

A simple way to explain:

| Framework | Best Mental Model |
|---|---|
| OpenAI SDK | Build custom Agents |
| CrewAI | Build Agent teams |
| LangGraph | Build stateful Agent workflows |

---

## Simple Progression

```text
LLM
 ↓
Single Agent
 ↓
Multi-Agent Team
 ↓
Stateful Workflow
 ↓
Enterprise AI Application
```

LangGraph becomes useful when workflows need:

- loops
- branching
- retries
- checkpoints
- state persistence
- human-in-the-loop review


# 12. Decision Guide

Ask these questions before choosing a framework.

## Question 1

Do I need maximum low-level control?

If yes, choose **OpenAI SDK**.

## Question 2

Do I want to quickly define Agent roles and tasks?

If yes, choose **CrewAI**.

## Question 3

Do I need complex stateful workflow with loops and checkpoints?

If yes, consider **LangGraph**.

## Question 4

Am I teaching faculty or students how Agents collaborate?

CrewAI is often easier to explain.


# 13. Classroom Discussion

There is no single correct answer.

The right choice depends on the problem.

# 16. Key Takeaways

In this bridge notebook, we learned:

- OpenAI SDK and CrewAI both support Agentic AI.
- OpenAI SDK gives more control over the workflow.
- CrewAI simplifies multi-agent collaboration.

- Framework choice depends on the problem, not popularity.
