````python
env_path = Path.cwd() / ".env"

print("Current working directory:", Path.cwd())
print("Looking for .env at:", env_path)
print("Env exists:", env_path.exists())

load_dotenv(env_path, override=True)
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    print("Dotenv keys found:", list(dotenv_values(env_path).keys()))
    raise ValueError("OPENAI_API_KEY not found. Please check your .env file.")

client = OpenAI(api_key=api_key)
print("OpenAI client initialized successfully.")
print("Key starts with:", api_key[:10], "...")
````

````python
def ask_openai(prompt, model="gpt-4o-mini", temperature=0.2, system_prompt="You are a helpful AI evaluation assistant.", max_tokens=800):
    response = client.chat.completions.create(
        model=model,
        temperature=temperature,
        max_tokens=max_tokens,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": prompt}
        ]
    )
    return response.choices[0].message.content
````

# 1️⃣ Why Evaluate AI?

Traditional software is mostly deterministic. AI systems are probabilistic.

```text
Traditional Software: Input → Logic → Expected Output
AI System: Prompt → LLM → Probabilistic Output
```

That is why AI needs systematic evaluation.

````python
comparison = pd.DataFrame([
    {"Aspect": "Traditional Software", "Nature": "Deterministic", "Testing": "Expected output comparison", "Risk": "Logic bug"},
    {"Aspect": "AI Systems", "Nature": "Probabilistic", "Testing": "Quality, safety and grounding metrics", "Risk": "Hallucination, drift, inconsistency"},
])
comparison
````

# 2️⃣ Prototype vs Production

Many AI demos look impressive. Production is different.

Problems appear when:
- many users ask different questions
- documents change
- prompts are modified
- models are updated
- edge cases appear
- hallucinations reach real users

````python
prototype_vs_production = pd.DataFrame([
    {"Stage": "Prototype", "Typical Behavior": "Works on demo examples", "Risk": "False confidence"},
    {"Stage": "Pilot", "Typical Behavior": "Works for selected users", "Risk": "Unseen scenarios"},
    {"Stage": "Production", "Typical Behavior": "Faces real-world variety", "Risk": "Trust loss if untested"},
])
prototype_vs_production
````

# 4️⃣ Golden Test Set

A Golden Test Set is a small collection of questions where you already know the correct answer.

Each item should include:
- question
- expected answer / ground truth
- expected source
- question type
- in-scope or out-of-scope

````python
golden_test_set = [
    {"question": "What is the M.Tech CSE fee per semester?", "ground_truth": "The tuition fee is Rs. 50,000 per semester.", "expected_source": "department_handbook.txt", "question_type": "factual", "scope": "in-scope"},
    {"question": "Who teaches the NLP elective?", "ground_truth": "Dr. Kavitha Raman coordinates the NLP elective.", "expected_source": "course_catalog.txt", "question_type": "factual", "scope": "in-scope"},
    {"question": "What is the hostel fee?", "ground_truth": "Not available in the provided knowledge base.", "expected_source": "none", "question_type": "out-of-scope", "scope": "out-of-scope"}
]
pd.DataFrame(golden_test_set)
````

# 5️⃣ Golden Test Set Checklist

A good set includes factual, procedural, comparative, temporal, out-of-scope, and ambiguous questions.

A focused 20–50 question set is better than a sloppy 200 question set.

````python
pd.DataFrame([
    {"Checklist Item": "Covers all important document sections", "Why": "Avoids blind spots"},
    {"Checklist Item": "Includes out-of-scope questions", "Why": "Tests refusal behavior"},
    {"Checklist Item": "Has known ground truth", "Why": "Enables objective scoring"},
    {"Checklist Item": "Includes different question types", "Why": "Tests retrieval breadth"},
    {"Checklist Item": "Updated continuously", "Why": "Reflects real user issues"},
])
````
