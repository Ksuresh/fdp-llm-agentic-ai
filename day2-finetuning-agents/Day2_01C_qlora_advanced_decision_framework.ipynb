# 1️⃣ QLoRA

QLoRA = Quantized LoRA.

It combines:
- 4-bit quantized base model
- trainable LoRA adapters
- low memory usage

This enables larger model fine-tuning on limited hardware.

````python
import pandas as pd
````

````python
qlora_table = pd.DataFrame([
    {"Approach": "Full FP32 Fine-tune", "VRAM": "112 GB", "Can Run On": "Multi-GPU cluster"},
    {"Approach": "Full FP16 Fine-tune", "VRAM": "56 GB", "Can Run On": "A100 GPU"},
    {"Approach": "LoRA FP16", "VRAM": "20 GB", "Can Run On": "RTX 3090 / 4090"},
    {"Approach": "QLoRA 4-bit + LoRA", "VRAM": "9 GB", "Can Run On": "Free Colab T4"},
])
qlora_table
````

# 2️⃣ QLoRA Configuration

````python
qlora_config_snippet = '''
from transformers import BitsAndBytesConfig
import torch

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
'''
print(qlora_config_snippet)
````

# 3️⃣ What 4-bit Quantization Means

Weights are stored in fewer bits to save memory.

The base model is quantized.  
LoRA adapters are trained.

Result:

> Lower memory + useful fine-tuning quality
