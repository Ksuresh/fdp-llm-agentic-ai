## What this notebook shows
An agent becomes useful only when it has tools. This notebook shows common tool categories using simple Python functions.

```python
import pandas as pd
import math
import textwrap

tool_taxonomy = pd.DataFrame([
    {"Tool Type": "Search / Retrieval", "Example": "Wikipedia / RAG", "Purpose": "Find knowledge"},
    {"Tool Type": "Code Execution", "Example": "Python calculator", "Purpose": "Compute exact answers"},
    {"Tool Type": "API Call", "Example": "Calendar / Gmail / GitHub", "Purpose": "Connect to systems"},
    {"Tool Type": "File Operation", "Example": "Read PDF / CSV / Excel", "Purpose": "Process documents"},
    {"Tool Type": "AI-as-Tool", "Example": "Summarizer / Translator", "Purpose": "Use another AI function"},
    {"Tool Type": "Human-in-the-loop", "Example": "Approval", "Purpose": "Control risky actions"}
])
tool_taxonomy
```

## Tool 1: Calculator
This is useful when the model should not guess numeric answers.

```python
def calculator(expression: str):
    try:
        return eval(expression, {"__builtins__": {}}, {"math": math})
    except Exception as e:
        return f"Error: {e}"

calculator("50000 * 4 * 0.5")
```

## Tool 2: Small Knowledge Search
This simulates a RAG/search tool.

```python
knowledge_base = {
    "rag": "RAG means Retrieval-Augmented Generation. It retrieves context before generating an answer.",
    "agent": "An AI agent can reason, choose tools, observe results, and continue until a goal is complete.",
    "mcp": "MCP stands for Model Context Protocol. It standardizes how models connect to tools and data.",
    "crewai": "CrewAI is used for role-based multi-agent collaboration.",
    "langgraph": "LangGraph is useful for stateful and controlled agent workflows."
}

def search_knowledge(query: str):
    query = query.lower()
    matches = []
    for key, value in knowledge_base.items():
        if key in query or any(word in value.lower() for word in query.split()):
            matches.append(value)
    return "\n".join(matches) if matches else "No matching knowledge found."

search_knowledge("What is RAG?")
```

## Tool 3: Human Approval
This is needed before risky actions like sending emails or updating records.

```python
def human_approval(action_description: str, approve=False):
    if approve:
        return f"Approved: {action_description}"
    return f"Blocked until human approval: {action_description}"

print(human_approval("Send marks email to 60 students", approve=False))
print(human_approval("Generate draft answer key", approve=True))
```

## Simple Agent Planner
The planner chooses which tool to use.

```python
def simple_agent(question: str):
    q = question.lower()

    if any(op in q for op in ["+", "-", "*", "/", "calculate", "total"]):
        expression = q.replace("calculate", "").replace("total", "").strip()
        return {
            "tool_used": "calculator",
            "answer": calculator(expression)
        }

    if "send" in q or "email" in q or "update" in q:
        return {
            "tool_used": "human_approval",
            "answer": human_approval(question, approve=False)
        }

    return {
        "tool_used": "search_knowledge",
        "answer": search_knowledge(question)
    }

questions = [
    "What is RAG?",
    "Calculate 50000 * 4 * 0.5",
    "Send marks email to students",
    "What is LangGraph?"
]

for q in questions:
    print("=" * 70)
    print("Question:", q)
    print(simple_agent(q))
```

## Key takeaway
Tools define what the agent can do. Missing tool means missing capability.
