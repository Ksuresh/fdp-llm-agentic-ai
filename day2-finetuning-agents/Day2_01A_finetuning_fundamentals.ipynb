# 1️⃣ Why Fine-Tuning?

In Day 1, we learnt:
- Prompt Engineering improves instructions.
- RAG adds private or changing knowledge.
- Multimodal AI handles images, documents and audio.

Now we ask:

> Can we change the model's behaviour itself?

Fine-tuning is useful when we need consistent behaviour, style, tone or output format.

````python
import pandas as pd
````

# 2️⃣ Decision Ladder

Always start simple.

```text
Prompt Engineering → RAG → Fine-Tuning → Pre-training
```

Never start with fine-tuning unless simpler methods fail.

# 3️⃣ RAG vs Fine-Tuning

| RAG | Fine-Tuning |
|---|---|
| Adds external knowledge | Changes behaviour |
| Reads documents at query time | Learns from examples |
| Easy to update | Needs retraining |
| Best for facts and policies | Best for style, tone, format |

Example: If syllabus changes tomorrow, use **RAG**, not fine-tuning.

# 4️⃣ Full Fine-Tuning vs PEFT

Full fine-tuning updates all model parameters.

PEFT = Parameter Efficient Fine-Tuning.  
It updates only a small number of additional parameters.

LoRA is one of the most popular PEFT techniques.

````python
vram_table = pd.DataFrame([
    {"Model": "FLAN-T5 Small", "Parameters": "80M", "Full FT": "0.3 GB", "LoRA": "0.05 GB", "QLoRA": "-", "Can Run On": "Laptop/CPU"},
    {"Model": "Mistral 7B", "Parameters": "7B", "Full FT": "28 GB", "LoRA": "14 GB", "QLoRA": "6 GB", "Can Run On": "Free Colab T4"},
    {"Model": "LLaMA 3 70B", "Parameters": "70B", "Full FT": "280 GB", "LoRA": "40 GB", "QLoRA": "20 GB", "Can Run On": "A100 GPUs"},
])
vram_table
````

# 5️⃣ LoRA Intuition

LoRA = Low-Rank Adaptation.

Instead of updating the full weight matrix:

```text
W
```

LoRA freezes W and learns a small update:

```text
W' = W + B × A
```

Only A and B are trained.

````python
lora_table = pd.DataFrame([
    {"Model": "FLAN-T5 Small", "Full Parameters": "80M", "LoRA Trainable": "0.26M", "Saved": "99.7%"},
    {"Model": "Mistral 7B", "Full Parameters": "7B", "LoRA Trainable": "19M", "Saved": "99.7%"},
    {"Model": "LLaMA 3 70B", "Full Parameters": "70B", "LoRA Trainable": "84M", "Saved": "99.9%"},
])
lora_table
````

# 6️⃣ What Fine-Tuning Learns

Fine-tuning is good for:
- response format
- style and tone
- domain vocabulary
- consistency
- instruction-following pattern

It is not ideal for frequently changing facts.  
Use RAG for that.

# 7️⃣ AI Mythbusters

## Myth
Fine-tuning teaches the model new facts.

## Reality
Fine-tuning mostly teaches behaviour and patterns.

Facts belong in RAG.  
Behaviour belongs in fine-tuning.

# 8️⃣ Dataset Quality

Quality matters more than quantity.

A fine-tuning example should be:
- accurate
- consistent
- well-formatted
- representative of real use
- reviewed by a human expert

````python
dataset_guidelines = pd.DataFrame([
    {"Rule": "Quality > Quantity", "Why": "50 excellent examples beat 500 noisy ones"},
    {"Rule": "Consistent Format", "Why": "The model learns the structure"},
    {"Rule": "Diverse Scenarios", "Why": "Improves generalization"},
    {"Rule": "Clean Responses", "Why": "Bad examples teach bad output"},
    {"Rule": "Human Review", "Why": "Training data becomes model behaviour"},
])
dataset_guidelines
````
