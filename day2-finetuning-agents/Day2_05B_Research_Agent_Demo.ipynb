# 05B Research Agent Demo

**Objective:** Build a simple transparent research agent with search and Python tools.

Day 2 · Session 5 · Agent Ecosystem

## What this notebook shows
A research agent can use more than one tool: search knowledge, calculate, and synthesize.

```python
import math
import textwrap
import pandas as pd

research_knowledge = {
    "transformer": '''
A Transformer is a neural network architecture based on self-attention.
It was introduced to process sequences efficiently and is widely used in modern LLMs.
Key ideas include attention, positional encoding, encoder-decoder blocks, and parallel training.
''',
    "bert": '''
BERT is an encoder-only Transformer model designed for understanding language.
It is commonly used for classification, question answering, and text understanding tasks.
''',
    "gpt": '''
GPT models are decoder-only Transformer models designed for text generation.
They predict the next token and are widely used for chatbots, writing, coding, and reasoning tasks.
''',
    "reinforcement learning": '''
Reinforcement learning is a machine learning paradigm where an agent learns by interacting with an environment.
It receives rewards or penalties and improves its policy over time.
'''
}

def wiki_like_search(query: str):
    q = query.lower()
    results = []
    for key, value in research_knowledge.items():
        if key in q:
            results.append(value.strip())
    return "\n\n".join(results) if results else "No reliable source found in local demo knowledge."

def python_calculator(expression: str):
    try:
        return eval(expression, {"__builtins__": {}}, {"math": math})
    except Exception as e:
        return f"Calculation error: {e}"
```

## Research Agent
This is a transparent, classroom-friendly agent. It prints Reason → Act → Observe.

```python
def research_agent(question: str):
    q = question.lower()
    print("User question:", question)
    print()

    if "perplexity" in q:
        print("Reason: This is a calculation question. I should use Python.")
        print("Act: python_calculator")
        # Perplexity for 10 tokens where each probability is 0.1:
        # PPL = exp(- average log probability)
        result = python_calculator("math.exp(-1 * math.log(0.1))")
        print("Observe:", result)
        return f"The perplexity is {result:.2f}."

    if "compare" in q and "gpt" in q and "bert" in q:
        print("Reason: I need information about both GPT and BERT.")
        print("Act: wiki_like_search('GPT')")
        gpt_info = wiki_like_search("gpt")
        print("Observe:", gpt_info[:120], "...")
        print()
        print("Act: wiki_like_search('BERT')")
        bert_info = wiki_like_search("bert")
        print("Observe:", bert_info[:120], "...")
        return (
            "GPT is decoder-only and optimized for generation. "
            "BERT is encoder-only and optimized for language understanding. "
            "Both are based on the Transformer architecture."
        )

    print("Reason: This is a concept question. I should search knowledge.")
    print("Act: wiki_like_search")
    observation = wiki_like_search(question)
    print("Observe:", observation)
    return observation

print(research_agent("What is a transformer neural network?"))
```

## Try more questions

```python
questions = [
    "Compare GPT and BERT",
    "Calculate perplexity of a model assigning probability 0.1 to each token",
    "Summarize reinforcement learning",
    "What is quantum networking?"
]

for q in questions:
    print("=" * 90)
    answer = research_agent(q)
    print("\nFinal answer:", answer)
    print()
```

## Show the agent trace as a table

```python
def trace_research_agent(question: str):
    trace = []
    q = question.lower()

    if "compare" in q and "gpt" in q and "bert" in q:
        trace.append({"Step": 1, "Reason": "Need GPT information", "Tool": "wiki_like_search", "Observation": wiki_like_search("gpt")[:80]})
        trace.append({"Step": 2, "Reason": "Need BERT information", "Tool": "wiki_like_search", "Observation": wiki_like_search("bert")[:80]})
        trace.append({"Step": 3, "Reason": "Synthesize comparison", "Tool": "LLM reasoning", "Observation": "GPT = generation, BERT = understanding"})
    else:
        trace.append({"Step": 1, "Reason": "Search concept", "Tool": "wiki_like_search", "Observation": wiki_like_search(question)[:80]})

    return pd.DataFrame(trace)

trace_research_agent("Compare GPT and BERT")
```
