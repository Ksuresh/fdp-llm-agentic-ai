## 1. Chatbot vs Agent
A chatbot answers. An agent plans, uses tools, observes results, and continues.

```python
import pandas as pd
import time
import json

comparison = pd.DataFrame({
    "Chatbot": [
        "Answers questions",
        "Usually one-turn",
        "Does not decide next action",
        "No tool execution by default",
        "Goal: provide information"
    ],
    "Agent": [
        "Achieves goals",
        "Multi-step",
        "Decides next action",
        "Uses tools and resources",
        "Goal: complete task"
    ]
})

comparison
```

## 3. Agent Components

```python
agent_parts = pd.DataFrame([
    {"Component": "Goal", "Meaning": "What the user wants to achieve"},
    {"Component": "LLM / Brain", "Meaning": "Reasons and decides next step"},
    {"Component": "Tools", "Meaning": "Functions/APIs the agent can use"},
    {"Component": "Resources", "Meaning": "Read-only knowledge/context"},
    {"Component": "Memory", "Meaning": "Stores previous steps or conversation"},
    {"Component": "Observation", "Meaning": "Result returned after action"},
])
agent_parts
```

## 4. Simple Agent Loop Simulation

```python
steps = [
    "Goal received",
    "Thinking about next step",
    "Selecting tool",
    "Executing tool",
    "Observing result",
    "Checking if goal is complete",
    "Final answer"
]

for step in steps:
    print("->", step)
    time.sleep(0.5)
```
