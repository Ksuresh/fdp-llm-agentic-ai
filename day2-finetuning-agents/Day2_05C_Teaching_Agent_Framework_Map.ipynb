## What this notebook shows
The same agent pattern can become a teaching assistant for any subject by changing only configuration.

```python
import pandas as pd

SUBJECT_CONFIG = {
    "subject": "Machine Learning",
    "level": "3rd year B.Tech CSE",
    "teaching_style": "Simple, Socratic, Indian classroom examples",
    "topic_focus": "Supervised learning, neural networks, evaluation metrics"
}

SUBJECT_CONFIG
```

## Subject Knowledge
For demo stability, we use local subject knowledge instead of live APIs.

```python
subject_knowledge = {
    "Machine Learning": {
        "overfitting": "Overfitting happens when a model performs well on training data but poorly on unseen data.",
        "precision": "Precision measures how many predicted positives are actually positive.",
        "recall": "Recall measures how many actual positives were correctly identified.",
        "gradient descent": "Gradient descent is an optimization method that updates parameters to reduce loss."
    },
    "Operating Systems": {
        "deadlock": "A deadlock occurs when processes wait forever for resources held by each other.",
        "paging": "Paging divides memory into fixed-size pages and frames.",
        "process": "A process is a program in execution."
    },
    "Data Structures": {
        "stack": "A stack is a LIFO data structure.",
        "queue": "A queue is a FIFO data structure.",
        "tree": "A tree is a hierarchical data structure with nodes and edges."
    }
}

def subject_search(subject, query):
    data = subject_knowledge.get(subject, {})
    q = query.lower()
    for key, value in data.items():
        if key in q:
            return value
    return "I don't know from the configured subject knowledge."
```

## Teaching Agent

```python
def teaching_agent(question, config):
    subject = config["subject"]
    level = config["level"]
    style = config["teaching_style"]

    raw_answer = subject_search(subject, question)

    if raw_answer.startswith("I don't know"):
        return raw_answer

    answer = f'''
Subject: {subject}
Level: {level}
Teaching style: {style}

Simple Explanation:
{raw_answer}

Classroom Example:
Imagine explaining this to a student during a lab session. First define the term, then show one small example.

Check Understanding:
Can you give one situation where this concept would be useful?
'''
    return answer.strip()

print(teaching_agent("What is overfitting?", SUBJECT_CONFIG))
```

## Change the subject in 5 lines

```python
SUBJECT_CONFIG = {
    "subject": "Operating Systems",
    "level": "2nd year B.Tech CSE",
    "teaching_style": "Simple explanation with real-world analogy",
    "topic_focus": "Processes, memory, scheduling, deadlocks"
}

print(teaching_agent("Explain deadlock", SUBJECT_CONFIG))
```

## Another example

```python
SUBJECT_CONFIG = {
    "subject": "Data Structures",
    "level": "1st year B.Tech CSE",
    "teaching_style": "Simple, beginner-friendly, with examples",
    "topic_focus": "Stacks, queues, linked lists, trees"
}

print(teaching_agent("What is a stack?", SUBJECT_CONFIG))
```
