````python
import os
from pathlib import Path
from dotenv import load_dotenv, dotenv_values
from openai import OpenAI
import pandas as pd
import json
````

````python
env_path = Path.cwd() / ".env"
print("Current working directory:", Path.cwd())
print("Looking for .env at:", env_path)
print("Env exists:", env_path.exists())

load_dotenv(env_path, override=True)
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    print("Dotenv keys found:", list(dotenv_values(env_path).keys()))
    raise ValueError("OPENAI_API_KEY not found. Please check your .env file.")

client = OpenAI(api_key=api_key)
print("OpenAI client initialized successfully.")
````

````python
def ask_openai(prompt, model="gpt-4o-mini", temperature=0.3,
               system_prompt="You are a helpful AI teaching assistant.",
               max_tokens=800):
    response = client.chat.completions.create(
        model=model,
        temperature=temperature,
        max_tokens=max_tokens,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": prompt}
        ]
    )
    return response.choices[0].message.content
````

````python
pd.DataFrame([
    {"Question": "Is the task clear?", "Fix": "State exactly what output is needed"},
    {"Question": "Is the audience specified?", "Fix": "Mention student level/background"},
    {"Question": "Is context provided?", "Fix": "Add what the learner already knows"},
    {"Question": "Is output format specified?", "Fix": "Ask for bullets/table/JSON"},
    {"Question": "Are there too many tasks?", "Fix": "Split into smaller prompts"},
    {"Question": "Are examples needed?", "Fix": "Add 2-3 few-shot examples"},
    {"Question": "Are constraints clear?", "Fix": "Set length, tone, style, exclusions"}
])
````

# 3️⃣ Debug a Bad Prompt

````python
bad_prompt = "Make questions on AI"
print(ask_openai(bad_prompt))
````

````python
debugged_prompt = """
You are an examiner for a 3rd-year B.Tech AI course.

Create 5 MCQ questions on Introduction to Artificial Intelligence.

Requirements:
- Difficulty: Medium
- Each question must have 4 options
- Include correct answer
- Include one-line explanation
- Avoid ambiguous options
- Cover different subtopics: search, ML, neural networks, ethics, agents

Return as a numbered list.
"""
print(ask_openai(debugged_prompt, max_tokens=1200))
````

# 5️⃣ Faculty Toolkit Builder

````python
subject = "Machine Learning"
student_level = "2nd-year engineering students"

toolkit_prompt = f"""
Create a reusable prompt toolkit for a faculty member teaching {subject} to {student_level}.

Include 8 prompt templates:
1. Concept explainer
2. Analogy generator
3. MCQ generator
4. Assignment feedback
5. Viva question generator
6. Lab activity generator
7. Common misconception finder
8. Research summary assistant

For each template include:
- Template name
- Full reusable prompt
- When to use it
"""
print(ask_openai(toolkit_prompt, max_tokens=1800))
````
