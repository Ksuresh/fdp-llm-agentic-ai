````python
import os
from pathlib import Path
from dotenv import load_dotenv, dotenv_values
from openai import OpenAI
import pandas as pd
import json
````

````python
repo_root = Path.cwd()

while not (repo_root / ".git").exists() and repo_root != repo_root.parent:
    repo_root = repo_root.parent

env_path = repo_root / ".env"

print("Repository root:", repo_root)
print("Loading .env from:", env_path)
print("Env exists:", env_path.exists())

load_dotenv(env_path, override=True)

api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    raise ValueError(f"OPENAI_API_KEY not found in {env_path}")

client = OpenAI(api_key=api_key)

print("OpenAI client initialized successfully.")
````

````python
def ask_openai(
    prompt,
    model="gpt-4o-mini",
    temperature=0.3,
    system_prompt="You are a helpful AI teaching assistant.",
    max_completion_tokens=800
):
    response = client.chat.completions.create(
        model=model,
        temperature=temperature,
        max_completion_tokens=max_completion_tokens,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": prompt}
        ]
    )

    return response.choices[0].message.content
````

# 1️⃣ Why Prompt Engineering Matters

The model remains the same, but the quality of the instruction changes.

A weak prompt often produces a generic response.  
A strong prompt improves relevance, structure, audience alignment and usability.

We will compare both using the same model.

````python
bad_prompt = "Explain machine learning"
print("WEAK PROMPT OUTPUT\n")
print(ask_openai(bad_prompt))
````

## 🔍 Observe

The answer may be correct, but it is likely to be generic.

The model does not know:

- Who the learner is
- What prior knowledge they have
- Why they need the explanation
- What output format is expected
- How detailed the answer should be

````python
good_prompt = """
Explain machine learning to a third-year B.Tech CSE student
who knows Python but has not studied machine learning yet.

Use a student exam-preparation analogy.

Requirements:
- Keep the answer under 120 words
- Avoid heavy mathematical terminology
- End with one check-your-understanding question
"""

print("IMPROVED PROMPT OUTPUT\n")
print(ask_openai(good_prompt))
````

# 2️⃣ Anatomy of an Effective Prompt

A strong prompt usually contains some or all of these elements:

- **Persona** — Who should the model act as?
- **Task** — What exactly should it do?
- **Context** — What background does it need?
- **Format** — How should the answer be presented?
- **Examples** — What does a good output look like?
- **Constraints** — What limits or rules must it follow?

Not every prompt needs all six elements, but adding the right ones improves precision and usefulness.

````python
prompt_anatomy = pd.DataFrame([
    {
        "Element": "Persona",
        "Purpose": "Defines the role, expertise or tone",
        "Example": "You are a friendly machine learning professor."
    },
    {
        "Element": "Task",
        "Purpose": "Defines the action to perform",
        "Example": "Explain overfitting."
    },
    {
        "Element": "Context",
        "Purpose": "Provides relevant background",
        "Example": "The learner understands train-test split."
    },
    {
        "Element": "Format",
        "Purpose": "Specifies the output structure",
        "Example": "Return the answer as five bullet points."
    },
    {
        "Element": "Examples",
        "Purpose": "Demonstrates the expected style or pattern",
        "Example": "Use an analogy similar to exam preparation."
    },
    {
        "Element": "Constraints",
        "Purpose": "Sets boundaries and limitations",
        "Example": "Use fewer than 150 words and avoid equations."
    }
])

prompt_anatomy
````

# 3️⃣ Prompt Evolution: Weak → Better → Expert

We will now improve the same prompt in three stages.

The model remains unchanged.

Only the clarity and completeness of the instruction improve.

````python
topic = "overfitting"

prompts = {
    "Weak": f"Explain {topic}.",

    "Better": f"""
Explain {topic} to a second-year engineering student
using simple language and one practical analogy.
""",

    "Expert": f"""
You are a machine learning professor teaching second-year engineering students.

Task:
Explain {topic}.

Context:
The students know Python and understand train-test split,
but they have not studied model generalization.

Requirements:
- Use a student exam-preparation analogy
- Explain why overfitting happens
- Mention one way to detect it
- Mention two ways to reduce it
- Keep the answer under 180 words
- End with one check-your-understanding question
"""
}

for level, prompt in prompts.items():
    print("=" * 80)
    print(level.upper(), "PROMPT OUTPUT")
    print("=" * 80)
    print(ask_openai(prompt))
    print()
````

# 4️⃣ Persona Experiment

````python
question = "Explain neural networks."
personas = {
    "Professor": "You are a senior AI professor. Explain clearly and academically.",
    "Friendly Tutor": "You are a friendly tutor. Use simple analogies.",
    "Strict Examiner": "You are a strict university examiner. Be precise and exam-oriented.",
    "Industry Mentor": "You are an industry AI mentor. Focus on practical use cases."
}

for name, system_prompt in personas.items():
    print("\n" + "="*80)
    print("Persona:", name)
    print("="*80)
    print(ask_openai(question, system_prompt=system_prompt))
````

# 5️⃣ Format and Constraints

````python
prompt_with_format = """
Compare supervised learning and unsupervised learning.

Return as a markdown table with columns:
- Aspect
- Supervised Learning
- Unsupervised Learning
- Example

Keep explanations short.
"""
print(ask_openai(prompt_with_format))
````
