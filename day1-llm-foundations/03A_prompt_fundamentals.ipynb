````python
import os
from pathlib import Path
from dotenv import load_dotenv, dotenv_values
from openai import OpenAI
import pandas as pd
import json
````

````python
env_path = Path.cwd() / ".env"
print("Current working directory:", Path.cwd())
print("Looking for .env at:", env_path)
print("Env exists:", env_path.exists())

load_dotenv(env_path, override=True)
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    print("Dotenv keys found:", list(dotenv_values(env_path).keys()))
    raise ValueError("OPENAI_API_KEY not found. Please check your .env file.")

client = OpenAI(api_key=api_key)
print("OpenAI client initialized successfully.")
````

````python
def ask_openai(prompt, model="gpt-4o-mini", temperature=0.3,
               system_prompt="You are a helpful AI teaching assistant.",
               max_tokens=800):
    response = client.chat.completions.create(
        model=model,
        temperature=temperature,
        max_tokens=max_tokens,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": prompt}
        ]
    )
    return response.choices[0].message.content
````

# 1️⃣ Why Prompt Engineering Matters

The model does not change. The prompt changes. A weak prompt gives a generic answer; a strong prompt gives a useful answer.

````python
bad_prompt = "Explain machine learning"
print(ask_openai(bad_prompt))
````

## 🔍 Observe

This answer is usually correct but generic. It may sound like a textbook.

````python
good_prompt = """
Explain machine learning to a 3rd-year B.Tech CSE student
who knows Python but has not studied ML yet.
Use a student exam-preparation analogy.
Keep it under 120 words.
End with one check-your-understanding question.
"""
print(ask_openai(good_prompt))
````

# 2️⃣ Anatomy of a Prompt

Persona, Task, Context, Format, Examples and Constraints.

````python
pd.DataFrame([
    {"Element": "Persona", "Use": "Defines role/style", "Example": "You are a friendly ML professor"},
    {"Element": "Task", "Use": "Defines action", "Example": "Explain overfitting"},
    {"Element": "Context", "Use": "Gives background", "Example": "Student knows train/test split"},
    {"Element": "Format", "Use": "Shapes answer", "Example": "Use a markdown table"},
    {"Element": "Examples", "Use": "Shows desired pattern", "Example": "Example input -> output"},
    {"Element": "Constraints", "Use": "Sets boundaries", "Example": "Avoid equations"}
])
````

# 3️⃣ Prompt Evolution: Bad → Better → Expert

````python
topic = "overfitting"
prompts = {
    "Beginner": f"Explain {topic}",
    "Better": f"Explain {topic} to a second-year engineering student in simple language.",
    "Expert": f"""
You are an ML professor teaching second-year engineering students.

Task: Explain {topic}.
Context: Students know Python and train/test split, but not regularization.
Format:
1. One-line definition
2. Simple analogy
3. One ML example
4. One common mistake
5. One question to test understanding

Constraints:
- Avoid heavy mathematics
- Keep under 180 words
"""
}

for label, prompt in prompts.items():
    print("\n" + "="*80)
    print(label)
    print("="*80)
    print(ask_openai(prompt))
````

# 4️⃣ Persona Experiment

````python
question = "Explain neural networks."
personas = {
    "Professor": "You are a senior AI professor. Explain clearly and academically.",
    "Friendly Tutor": "You are a friendly tutor. Use simple analogies.",
    "Strict Examiner": "You are a strict university examiner. Be precise and exam-oriented.",
    "Industry Mentor": "You are an industry AI mentor. Focus on practical use cases."
}

for name, system_prompt in personas.items():
    print("\n" + "="*80)
    print("Persona:", name)
    print("="*80)
    print(ask_openai(question, system_prompt=system_prompt))
````

# 5️⃣ Format and Constraints

````python
prompt_with_format = """
Compare supervised learning and unsupervised learning.

Return as a markdown table with columns:
- Aspect
- Supervised Learning
- Unsupervised Learning
- Example

Keep explanations short.
"""
print(ask_openai(prompt_with_format))
````
