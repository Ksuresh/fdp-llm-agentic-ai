## 1. Why Do Agents Need Tools?
LLMs are excellent at reasoning, but they cannot directly access live systems like databases, APIs, emails, or files. Agents become useful by invoking **Tools** to perform these actions.

```
from pathlib import Path
from dotenv import load_dotenv
import os

from agents import Agent, Runner, function_tool
```

## 2. Load Environment Variables

```
current = Path.cwd().resolve()

for folder in [current] + list(current.parents):
    env = folder / ".env"
    if env.exists():
        load_dotenv(env)
        break

print("API Key Loaded:", os.getenv("OPENAI_API_KEY") is not None)
```

## 3. Create Your First Function Tool

```
@function_tool
def multiply(a: int, b: int) -> int:
    """
    Multiply two numbers.
    """
    return a * b
```

### What did we just learn?
- `@function_tool` exposes a Python function to the Agent.
- The docstring helps the LLM understand when to use it.
- Type hints describe the expected inputs.

## 4. Create an Agent

```
math_agent = Agent(
    name="Math Assistant",
    instructions="You are a helpful mathematics assistant.",
    tools=[multiply],
)
```

## 5. Run the Agent

```
result = Runner.run_sync(
    math_agent,
    "What is 27 multiplied by 53?"
)

print(result.final_output)
```

### Pause & Reflect
Notice that we never called `multiply()` directly. The Agent decided to invoke the tool.

## 6. Create Another Tool

```
@function_tool
def faculty_directory(department: str) -> str:
    """
    Returns faculty members for a department.
    """
    directory = {
        "AI&ML": "Prof. Arun, Prof. Meena",
        "CSE": "Prof. Kumar, Prof. Rekha",
        "ECE": "Prof. Ramesh, Prof. Divya"
    }

    return directory.get(department.upper(), "Department not found.")
```

## 7. Agent with Multiple Tools

```
college_agent = Agent(
    name="DSU Assistant",
    instructions="You help faculty answer questions about the college.",
    tools=[multiply, faculty_directory],
)
```

```
response = Runner.run_sync(
    college_agent,
    "Who teaches in AI&ML?"
)

print(response.final_output)

response = Runner.run_sync(
    college_agent,
    "Calculate 84 multiplied by 19."
)

print(response.final_output)
```
