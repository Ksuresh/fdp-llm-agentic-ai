## Step 1 — Import Libraries

We only need three things:

- `Agent`: defines the AI assistant
- `Runner`: runs the agent
- `dotenv`: loads the API key from `.env`


````python
from agents import Agent, Runner
from dotenv import load_dotenv
from pathlib import Path
import os

print("Imports successful")
````

## Step 2 — Load API Key from Root `.env`

Your project structure is:

```text
FDP-LLM/
│
├── .env
├── requirements.txt
├── day3-openai-crewai/
│   └── Day3_01A_Hello_OpenAI_Agents_SDK.ipynb
```

This cell searches upward from the current folder until it finds `.env`.


````python
def load_root_env() -> Path:
    """Search current and parent folders for .env and load it."""
    current = Path.cwd().resolve()

    for folder in [current, *current.parents]:
        env_path = folder / ".env"
        if env_path.exists():
            load_dotenv(env_path)
            return env_path

    raise FileNotFoundError(
        "Could not find .env. Please keep .env in the project root folder."
    )


env_path = load_root_env()

if not os.getenv("OPENAI_API_KEY"):
    raise ValueError("OPENAI_API_KEY not found inside .env")

print(f"API key loaded from: {env_path}")
````

## Step 3 — Create Your First Agent

An **Agent** is an LLM with a role, goal, and instructions.

For this FDP, we will use a running example:

> **DSU AI Teaching Assistant**


````python
student_assistant = Agent(
    name="DSU AI Teaching Assistant",
    instructions="""
You are a helpful AI teaching assistant for engineering professors and students.

Your teaching style:
- Explain concepts in simple English
- Use Indian engineering college examples
- Avoid unnecessary jargon
- Keep answers crisp and classroom-friendly
- Use bullet points only when helpful
""",
    model="gpt-4o-mini",
)

print("Agent created:", student_assistant.name)
````

## Step 4 — Run the Agent

In notebooks, use:

```python
await Runner.run(...)
```

In a normal `.py` script, you may use:

```python
Runner.run_sync(...)
```

Since we are using VS Code notebooks, we will use `await Runner.run()`.


````python
result = await Runner.run(
    student_assistant,
    "Explain machine learning in 3 simple points."
)

print(result.final_output)
````

## Step 5 — Ask More Questions

The same agent can answer many questions while following the same instructions.


````python
questions = [
    "Explain overfitting with a simple classroom example.",
    "Explain gradient descent in two sentences.",
    "What is the difference between AI, ML, and Deep Learning?"
]

for question in questions:
    print("\n" + "=" * 80)
    print("QUESTION:", question)
    print("=" * 80)

    result = await Runner.run(student_assistant, question)
    print(result.final_output)
````

## Step 6 — Same Model, Different Agents

Here we use the **same model** but different instructions.

This proves an important idea:

> The model is the engine. The agent instructions define the behavior.


````python
strict_professor = Agent(
    name="Strict Professor",
    instructions="""
You are a strict university professor.

Answer in exam style.
Use precise academic language.
Keep the answer concise.
Do not use casual examples.
""",
    model="gpt-4o-mini",
)

friendly_teacher = Agent(
    name="Friendly Teacher",
    instructions="""
You are a friendly teacher.

Use simple analogies.
Make the answer easy for a first-year engineering student.
Keep the tone warm and encouraging.
""",
    model="gpt-4o-mini",
)

print("Both agents created")
````

## Step 7 — Compare Agent Behavior

Same question. Same model. Different instructions.


````python
question = "Explain supervised learning."

strict_result = await Runner.run(strict_professor, question)
friendly_result = await Runner.run(friendly_teacher, question)

print("STRICT PROFESSOR OUTPUT")
print("-" * 80)
print(strict_result.final_output)

print("\n\nFRIENDLY TEACHER OUTPUT")
print("-" * 80)
print(friendly_result.final_output)
````
