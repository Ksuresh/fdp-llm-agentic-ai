# 1. Start with the Pain, Not the Protocol

Many people explain MCP like this:

> MCP is a protocol that connects AI applications to external tools and data sources.

That is technically correct, but for teaching, it is better to start with the problem.

LLMs do not automatically know your live systems. They do not know your college timetable, library hours, faculty directory, lab availability, course catalog, academic calendar, ERP data, or internal policies.

So if an AI assistant must answer real questions, it needs to connect to real systems.


# 2. DSU Smart Campus Assistant Story

Imagine we are building a DSU Smart Campus Assistant.

A professor asks:

> What is my first class today?

The AI needs access to the timetable.

Then the professor asks:

> Is the library open now?

The AI needs library information.

Then:

> Show me the AI elective syllabus.

The AI needs the course catalog.

Then:

> Is Lab 5 available tomorrow afternoon?

The AI needs lab booking information.

## Key Point

The LLM alone cannot answer these reliably. It needs a standard way to connect to campus systems.


# 3. Without MCP: The Integration Mess

Without MCP, every AI application may build its own connectors.

```text
Faculty Assistant
      |
      |-- Custom Timetable API Wrapper
      |-- Custom Library API Wrapper
      |-- Custom Course Catalog Wrapper
      |-- Custom Lab Booking Wrapper
      |-- Custom Academic Calendar Wrapper
```

Now suppose tomorrow we use another framework:

```text
OpenAI SDK
CrewAI
LangGraph
Claude Desktop
VS Code Agent
Cursor
```

Each one may need its own integration code. That becomes difficult to maintain.


# 4. Same Campus Data, Many Frameworks

Assume we have one system:

```text
Faculty Directory
```

Without a standard protocol, each AI tool needs a separate connector.

```text
OpenAI Agent       -> Faculty Directory Wrapper 1
CrewAI Agent       -> Faculty Directory Wrapper 2
LangGraph Workflow -> Faculty Directory Wrapper 3
Claude Desktop     -> Faculty Directory Wrapper 4
VS Code Agent      -> Faculty Directory Wrapper 5
```

Same data. Repeated integration effort.

This is the pain MCP tries to solve.


# 5. With MCP: One Server, Many Clients

With MCP, the system exposes one standard MCP server.

```text
Faculty Directory
        |
        v
Faculty Directory MCP Server
        |
        |-- OpenAI Agent
        |-- CrewAI Agent
        |-- LangGraph Workflow
        |-- Claude Desktop
        |-- VS Code Agent
        |-- Other AI Applications
```

Now the integration is reusable.

The AI framework may change. The system integration remains standard.

## Big Idea

> Frameworks change. Protocols stay.


# 6. USB Analogy

Before USB, many devices had different cables.

```text
Printer  -> Special cable
Mouse    -> Different cable
Keyboard -> Different cable
Scanner  -> Different cable
Camera   -> Different cable
```

USB created a common connection standard.

```text
USB Port
   |-- Mouse
   |-- Keyboard
   |-- Printer
   |-- Scanner
   |-- Camera
```

MCP plays a similar role for AI integrations.

> MCP is like USB for AI applications.


# 7. Why Tool Calling Alone Is Not Enough

Tool calling lets an Agent call a function. That is useful.

But tool calling is usually defined inside one framework or one application.

```text
OpenAI Function Tool
CrewAI Tool
LangGraph Tool Node
Claude Tool
Custom Python Wrapper
```

All may look different.

MCP standardizes how tools and context are exposed to AI applications.

| Tool Calling | MCP |
|---|---|
| Usually framework-specific | Protocol-based |
| Often defined inside one app | Reusable across apps |
| Good for local tools | Good for shared systems |
| Developer writes wrappers repeatedly | One server can serve many clients |


# 9. Important Vocabulary

We will study these in the next notebook.

| Term | Simple Meaning |
|---|---|
| Host | The AI application the user interacts with |
| Client | The component that talks MCP |
| Server | Exposes tools, resources, and prompts |
| Tool | An action the AI can ask the server to perform |
| Resource | Read-only information the AI can access |
| Prompt | Reusable instruction/template exposed by the server |
| Transport | How client and server communicate |

For today, remember:

> MCP Server exposes capabilities. MCP Client consumes them.


# 10. DSU Smart Campus: What Should Be Exposed?

| Campus System | Why AI Needs It |
|---|---|
| Faculty Directory | Find faculty names, departments, coordinators |
| Library | Check timings, rules, services |
| Course Catalog | Find subjects, credits, outcomes |
| Academic Calendar | Check events, holidays, exams |
| Lab Booking | Check lab availability |
| FDP Registry | Find upcoming FDPs and registrations |
| Placement Office | Find placement training schedule |

All of these can be exposed through MCP in a standard way.


# 11. Tool or Resource? Quick Preview

MCP is not only about tools. It also has resources.

| Campus Feature | Tool or Resource? | Why |
|---|---|---|
| Library timings | Resource | Read-only information |
| Faculty directory | Resource | Read-only information |
| Course catalog | Resource | Read-only information |
| Book a lab | Tool | Performs an action |
| Register for FDP | Tool | Changes system state |
| Send email | Tool | Performs an operation |

We will go deeper into this later.


# 12. Where MCP Fits in Our FDP

| Technology | Main Responsibility |
|---|---|
| LLM | Generate and reason with language |
| OpenAI SDK | Build Agents |
| CrewAI | Coordinate Agent teams |
| LangGraph | Control stateful workflows |
| MCP | Connect AI to external systems |

## Final Progression

```text
LLMs answer questions.
Agents perform tasks.
Workflows coordinate steps.
MCP connects AI to the real world.
```


# 13. Enterprise Mapping

A banking assistant needs customer profile, account balance, loan status, KYC status, transactions, fraud signals, and complaint tickets.

```text
DevOps MCP Server
      |-- Log Resource
      |-- Metrics Resource
      |-- Deployment Tool
      |-- Incident Ticket Tool
      |-- Runbook Resource
```

# 16. Common Misunderstandings

Avoid these misunderstandings:

1. MCP is not an LLM.
2. MCP is not an Agent framework.
3. MCP does not replace OpenAI SDK, CrewAI, or LangGraph.
4. MCP is not only for Claude.
5. MCP is not only for tools.
6. MCP is a protocol for connecting AI applications to external context and capabilities.

## Instructor Tip

Repeat this sentence:

> MCP is not the brain. MCP is the connector.
