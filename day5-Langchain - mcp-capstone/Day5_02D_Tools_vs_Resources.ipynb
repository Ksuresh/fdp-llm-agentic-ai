# 1. Why This Notebook Matters

Many beginners think:

> MCP = Tools

That is incomplete.

MCP can expose:

```text
Tools
Resources
Prompts
```

This notebook focuses on the most common confusion:

```text
Tool vs Resource
```

Understanding this distinction is important for designing clean, safe, and scalable MCP systems.


# 2. Simple Rule

Use this simple rule:

```text
Do you need to READ information?
        |
        v
    Resource

Do you need to DO something?
        |
        v
      Tool
```

# 4. Resource Mental Model

A Resource is like a reference book.

You read it.

You do not change it.

```text
Library timings
Course catalog
Faculty directory
Academic calendar
Policy documents
Exam schedule
```

Resources are useful when the AI needs context before answering.


# 5. Tool Mental Model

A Tool is like a service counter.

You ask it to perform an operation.

```text
Book lab
Register FDP
Send email
Create ticket
Cancel booking
Update attendance
```

Tools are more powerful because they can change the world.

That also makes them riskier.


# 6. Safety Difference

Resources are usually lower risk.

Tools can be high risk.

| MCP Type | Risk Level | Example |
|---|---|---|
| Resource | Usually lower | Read library timings |
| Tool | Can be higher | Cancel a lab booking |
| Tool | High | Update marks |
| Tool | Very high | Approve payment |

---

## Important Point

The more a capability changes real systems, the more guardrails you need.

# 11. Decision Framework

Use this decision framework when designing MCP systems.

```text
Step 1: Is the capability read-only?
    Yes -> Resource

Step 2: Does it perform an operation?
    Yes -> Tool

Step 3: Does it change records?
    Yes -> Tool with guardrails

Step 4: Does it require approval?
    Yes -> Tool + human-in-the-loop

Step 5: Is it reusable instruction?
    Yes -> Prompt
```

This helps avoid poor MCP design.


# 12. Code Preview: Tools vs Resources

The exact syntax may vary depending on the MCP SDK version.

Conceptually:

```python
@mcp.tool()
def book_lab(...):
    ...

@mcp.resource("campus://library/timings")
def library_timings():
    ...
```

Tools are functions the AI can call to perform actions.

Resources expose information that the AI can read.


# 13. Example Resource Design

A clean resource URI should be understandable.

Examples:

```text
campus://faculty/directory
campus://library/timings
campus://courses/catalog
campus://calendar/academic
campus://labs/availability
```

This makes the MCP Server easier to understand and maintain.


# 14. Example Tool Design

Tool names should describe actions clearly.

Good examples:

```text
book_lab
register_for_fdp
send_faculty_email
create_support_ticket
cancel_lab_booking
```

Poor examples:

```text
do_action
process_request
update_data
campus_tool
```

Names matter because the AI uses names and descriptions to decide what to call.

# 19. Why Not Make Everything a Tool?

Beginners often expose everything as tools.

This works for demos, but it is poor design.

Problems:

- Too many tools confuse the Agent.
- Read-only information becomes unnecessarily operational.
- Security design becomes harder.
- Maintenance becomes messy.
- It becomes unclear what changes state and what does not.

Better design:

```text
Read-only knowledge -> Resource
Action or mutation -> Tool
Reusable instruction -> Prompt
```


# 20. Why Not Make Everything a Resource?

The opposite mistake is also possible.

If a user says:

> Register me for the FDP

That is not a resource.

The system must perform an action.

So it should be a Tool.

A Resource can inform the Agent about FDP details.

A Tool can actually register the participant.

```text
Resource: fdp_list
Tool: register_fdp()
```


# 21. Prompt Preview

Prompts are the third important MCP primitive.

A prompt is reusable guidance.

Example:

```text
student_helpdesk_prompt
faculty_support_prompt
compliance_response_prompt
rca_summary_prompt
```

In DSU:

```text
Prompt: student_helpdesk_prompt
Rules:
- Use simple language
- Do not reveal confidential data
- Refer official office for final confirmation
```

We will keep prompts conceptual in this FDP.


# 22. Practical Campus Design

For DSU Smart Campus MCP Server, a good design may look like:

```text
Resources:
  campus://faculty/directory
  campus://library/timings
  campus://courses/catalog
  campus://calendar/academic
  campus://labs/availability

Tools:
  book_lab()
  register_fdp()
  send_email()
  create_helpdesk_ticket()

Prompts:
  student_helpdesk_prompt
  faculty_support_prompt
```

This is much cleaner than exposing everything as tools.

# 25. Common Mistakes

Avoid these mistakes:

1. Making every capability a tool.
2. Making every capability a resource.
3. Ignoring authorization for sensitive resources.
4. Allowing tools to change data without confirmation.
5. Using vague tool names.
6. Using unclear resource URIs.
7. Not separating read-only data from actions.
8. Forgetting audit logs for tools.
9. Exposing confidential data as public resources.
10. Not documenting capabilities clearly.
