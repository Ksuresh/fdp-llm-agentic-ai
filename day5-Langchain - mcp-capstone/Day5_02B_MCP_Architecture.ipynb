# 1. Big Picture Architecture

MCP has three main architectural layers:

```text
Host Application
      |
      v
MCP Client
      |
      v
MCP Server
      |
      |-- Tools
      |-- Resources
      |-- Prompts
      |
      v
External Systems
```

In the DSU Smart Campus Assistant:

```text
Smart Campus Assistant
      |
      v
MCP Client
      |
      v
DSU Campus MCP Server
      |
      |-- Faculty Directory
      |-- Library
      |-- Course Catalog
      |-- Academic Calendar
      |-- Lab Booking
```


# 2. What is an MCP Host?

The **Host** is the application where the user interacts with AI.

Examples:

- Claude Desktop
- ChatGPT / OpenAI Agent application
- VS Code / Cursor
- Custom enterprise AI assistant
- Smart Campus Assistant
- Banking AI assistant

In our running example:

```text
Professor
    |
    v
DSU Smart Campus Assistant
```

The DSU Smart Campus Assistant is the Host application.


# 3. What is an MCP Client?

The **MCP Client** lives inside or near the Host application.

It knows how to speak the MCP protocol.

Simple explanation:

> The Client is the messenger between the AI application and the MCP Server.

```text
Host Application
      |
      v
MCP Client
      |
      v
MCP Server
```

The user usually does not see the client directly.

It works behind the scenes.


# 4. What is an MCP Server?

The **MCP Server** exposes capabilities to AI applications.

It can expose:

- Tools
- Resources
- Prompts

It connects to real systems.

In DSU:

```text
DSU Campus MCP Server
      |
      |-- Faculty Directory
      |-- Library System
      |-- Course Catalog
      |-- Academic Calendar
      |-- Lab Booking System
```

The server is where institutional capabilities are made available in a standard way.


# 5. MCP Architecture with DSU Example

```text
Professor
   |
   v
DSU Smart Campus Assistant  [Host]
   |
   v
MCP Client
   |
   v
DSU Smart Campus MCP Server
   |
   |-- Tool: book_lab()
   |-- Tool: register_fdp()
   |-- Resource: faculty_directory
   |-- Resource: library_timings
   |-- Resource: course_catalog
   |-- Prompt: student_helpdesk_prompt
```

This architecture is clean because the AI assistant does not need separate custom integrations for every campus system.


# 6. MCP Tools

A **Tool** is something the AI can ask the server to perform.

Tools usually do actions.

Examples:

```text
book_lab()
register_for_fdp()
send_email()
create_ticket()
update_attendance()
submit_assignment()
```

In the Smart Campus Assistant:

```text
Professor asks:
"Book Lab 5 tomorrow from 2 PM to 4 PM."

Assistant uses:
book_lab()
```

## Simple Rule

If it performs an action, it is usually a Tool.


# 7. MCP Resources

A **Resource** is information the AI can read.

Resources are usually read-only.

Examples:

```text
faculty_directory
library_timings
course_catalog
academic_calendar
exam_schedule
placement_policy
```

In the Smart Campus Assistant:

```text
Professor asks:
"When is the library open?"

Assistant reads:
library_timings
```

## Simple Rule

If it is read-only information, it is usually a Resource.


# 8. MCP Prompts

A **Prompt** is a reusable instruction template exposed by the MCP Server.

Example:

```text
student_helpdesk_prompt
faculty_support_prompt
course_advisor_prompt
incident_summary_prompt
```

Why is this useful?

Because institutions can standardize how AI should respond.

For example, DSU could define:

```text
When answering students:
- Use simple language
- Do not reveal confidential data
- Refer official policies when required
- Ask students to contact office for final confirmation
```

Prompts help maintain consistency.


# 9. Tools vs Resources vs Prompts

| MCP Concept | Simple Meaning | Campus Example |
|---|---|---|
| Tool | Performs action | Book lab, register FDP, send email |
| Resource | Provides read-only data | Library timings, faculty directory |
| Prompt | Reusable instruction | Student helpdesk response template |

---

## Instructor Tip

Ask participants:

> Is "faculty directory" a tool or resource?

Answer: Resource.

> Is "send email to faculty" a tool or resource?

Answer: Tool.


# 10. Transport

Transport is how MCP Client and MCP Server communicate.

Simple meaning:

> Transport is the communication channel.

Examples may include:

- Local process communication
- Standard input/output
- HTTP-based communication
- Streamable communication depending on implementation

For beginners, do not worry about transport details.

Just remember:

```text
Client and Server need a communication channel.
That channel is called transport.
```


# 11. Sampling - High-Level View

Sampling is an advanced MCP concept.

At a high level, it allows an MCP Server to ask the Host/LLM to perform model completion in certain flows.

For this FDP, we will keep it conceptual only.

Why?

Because faculty first need to understand:

- Tools
- Resources
- Client
- Server

Sampling can be introduced later as an advanced topic.

## Simple Explanation

> Sampling is when the MCP flow may involve model generation through the host.

We will not implement sampling in this workshop.


# 12. End-to-End Request Flow

Let us trace a real request.

Professor asks:

> Who is the AI&ML department coordinator?

Flow:

```text
1. Professor asks question
2. Smart Campus Assistant receives it
3. MCP Client sends request to MCP Server
4. MCP Server exposes faculty directory resource
5. Assistant uses the returned information
6. Professor receives the answer
```

Architecture:

```text
Professor
   |
   v
Smart Campus Assistant
   |
   v
MCP Client
   |
   v
Campus MCP Server
   |
   v
Faculty Directory Resource
```


# 13. Another Request Flow: Action

Professor asks:

> Book Lab 5 tomorrow from 2 PM to 4 PM.

This is not just reading data.

It performs an action.

Flow:

```text
1. Professor asks to book lab
2. Assistant identifies action
3. MCP Client calls MCP Server
4. MCP Server exposes book_lab tool
5. Tool checks availability and books lab
6. Assistant confirms booking
```

Architecture:

```text
Professor
   |
   v
Smart Campus Assistant
   |
   v
MCP Client
   |
   v
Campus MCP Server
   |
   v
book_lab Tool
```


# 14. MCP and Security

MCP makes integration easier, but security is still critical.

For campus systems, we must consider:

- Who is asking?
- What are they allowed to access?
- Is the data confidential?
- Is the action read-only or write?
- Should human approval be required?
- Should logs be maintained?

Example:

Reading library timings is safe.

Changing exam marks is high risk.

Booking a lab may need authorization.


# 19. Common Mistakes

Avoid these misunderstandings:

1. Thinking MCP is only tool calling.
2. Ignoring resources.
3. Ignoring prompts.
4. Confusing Host and Client.
5. Exposing write actions without authorization.
6. Treating all data as safe.
7. Forgetting audit logs.
8. Building one-off integrations instead of reusable MCP servers.
